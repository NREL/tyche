```r
require(data.table)
require(ggplot2)
```

```r
z <- fread("results3.tsv", stringsAsFactors=TRUE)
```

```r
summary(z)
```

```r
g <- ggplot(z, aes(x=`Result`, color=`Factor 1`)) +
    geom_density() +
    labs(
        x="Total Cost [$/kg]",
        y="Probability Density [kg/$]",
        color="Factor",
        title="Two-Stage Stochastic Optimization of PV PolySilicon"
    ) +
    theme(legend.justification = c(1, 1), legend.position = c(1, 1))
g
```

```r
png("results3.png", width=6, height=7.5, units="in", res=300)
g
dev.off()
```

```r
pdf("results3.pdf", width=8.5, height=11)
g + theme(plot.margin = unit(c(0.75, 0.75, 0.75, 0.75), "in"))
dev.off()
```

```r
g <- ggplot(
    data.table(
        `Parameteric Improvement [%]`=100*(exp((0:10) * log(10) / 10) - 1),
        `Probability` = c(0.290, 0.025, 0.050, 0.075, 0.100, 0.105, 0.105, 0.100, 0.075, 0.050, 0.025)
    ), aes(x=`Parameteric Improvement [%]`, y =`Probability`)) +
    geom_bar(stat="identity")
g
```
